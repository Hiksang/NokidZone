```python
import openpyxl
import time
import pandas as pd

## 엑셀파일 열기
filename = "nokids.xlsx"
exelFile = openpyxl.load_workbook(filename)

## 시트 설정
sheet = exelFile.worksheets[0]

df = pd.DataFrame()
adr = []
rowCount = 1

for row in sheet.rows:
    try:
        a = row[0].value
        b = row[2].value
        c = row[3].value
        print(row[0].value)
        print(row[1].value)
        print(row[2].value)
        k = dict()
        k['title'] = a
        k['lat'] = b
        k['lng'] = c
        adr.append(k)
    except:
        print("error")

print(adr)
# exelFile.save(".xlsx")
```
